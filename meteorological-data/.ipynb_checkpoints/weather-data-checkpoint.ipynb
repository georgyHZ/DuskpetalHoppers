# Meterological Measurement Data
Dataset retrieved from https://www.knmi.nl/nederland-nu/klimatologie/daggegevens

All measurements taken near Deelen Airport, Netherlands

```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to your file
file_path = "hourly-MD_Deelen_2021-2023.txt"
```

## Load the weather data file, using the load_knmi_file() function
This removes the description from the data .txt file

```python
# Read while skipping comment lines
df = pd.read_csv(file_path, comment='#')

# Show the first few rows
df.info()
```

## Rename columns accordingly

```python
column_rename_map = {
    'STN': 'Station',
    'YYYYMMDD': 'Date',
    'HH': 'Hour',
    'DD': 'WindDirection',
    'FH': 'WindSpeedMean_10min',
    'FF': 'WindSpeedLast10Min',
    'FX': 'WindGustMax',
    'T': 'Temperature',
    'T10N': 'MinTemp_10cm_6hr',
    'TD': 'DewPointTemp',
    'SQ': 'SunshineDuration',
    'Q': 'GlobalRadiation',
    'DR': 'PrecipitationDuration',
    'RH': 'PrecipitationAmount',
    'P': 'AirPressure',
    'VV': 'Visibility',
    'N': 'CloudCover',
    'U': 'RelativeHumidity',
    'WW': 'WeatherCode',
    'IX': 'WeatherCodeIndicator',
    'M': 'Fog',
    'R': 'Rain',
    'S': 'Snow',
    'O': 'Thunder',
    'Y': 'IceFormation'
}
# Strip leading/trailing spaces from all column names
df.columns = df.columns.str.strip()

df.rename(columns=column_rename_map, inplace=True)
df.info()
```

```python
df.head(25)
```

## Change the Date and Hour column into a single Timestamp column

```python
# 1) Make sure Date and Hour are integers
df['Date'] = pd.to_numeric(df['Date'], errors='coerce').astype(int)
df['Hour'] = pd.to_numeric(df['Hour'], errors='coerce').astype(int)

# 2) Build midnight-of-date
midnight = pd.to_datetime(df['Date'], format='%Y%m%d')

# 3) Shift hours down by one (1→0, …, 24→23)
hour_offset = df['Hour'] - 1

# 4) Create Timestamp
df['Timestamp'] = midnight + pd.to_timedelta(hour_offset, unit='h')

# 5) (Optional) Make it your index and drop originals
df.set_index('Timestamp', inplace=True, drop=True)
df.drop(columns=['Date','Hour'], inplace=True)

df
```

## Weather data statistics 2023 overview

```python
df.info()
```

```python
# 1) Filter to 2023 using the DatetimeIndex
df_2023 = df[df.index.year == 2023]

# 2) Run describe() on the filtered set
summary_2023 = df_2023.describe()  

# 3) (Optional) Transpose for easier reading
summary_2023_T = summary_2023.T

# 4) Display
print(summary_2023_T)

df_2023.to_csv('../weather.csv')
```

# Visualizations 

```python
# 0) Prep: make a true mm column once
df['Precipitation_mm'] = df['PrecipitationAmount'].astype(float) / 10.0

# 1) Time‐series of hourly precipitation in 2023
df_2023 = df[df.index.year == 2023]

plt.figure(figsize=(14, 5))
plt.plot(
    df_2023.index,
    df_2023['Precipitation_mm'],
    label='Precipitation (mm/h)',
    linewidth=1
)
plt.title('Hourly Precipitation in 2023 (KNMI Station 275)')
plt.xlabel('Timestamp')
plt.ylabel('Precipitation (mm)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df_rain = df_2023[df_2023['Precipitation_mm'] >= 0.1]

plt.figure(figsize=(10, 5))
plt.hist(
    df_rain['Precipitation_mm'],
    bins=30,
    edgecolor='black',
    log='y'
)
plt.title('Distribution of Hourly Precipitation in 2023')
plt.xlabel('Precipitation (mm)')
plt.ylabel('Number of Hours')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 3) Cumulative daily precipitation: 2021 vs 2023
#    (align on day‐of‐year to compare the seasonal accumulation)
df_2021 = df[df.index.year == 2021].copy()
df_2023 = df_2023.copy()

# Resample to daily sums
daily_2021 = df_2021['Precipitation_mm'].resample('D').sum()
daily_2023 = df_2023['Precipitation_mm'].resample('D').sum()

# Compute cumulative sums
cum_2021 = daily_2021.cumsum()
cum_2023 = daily_2023.cumsum()

# Day‐of‐year index
doy_2021 = cum_2021.index.dayofyear
doy_2023 = cum_2023.index.dayofyear

plt.figure(figsize=(14, 6))
plt.plot(doy_2021, cum_2021, label='2021', linewidth=2)
plt.plot(doy_2023, cum_2023, label='2023', linewidth=2)

plt.title('Cumulative Daily Precipitation: 2021 vs 2023', fontsize=14)
plt.xlabel('Day of Year', fontsize=12)
plt.ylabel('Cumulative Precipitation (mm)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Year')
plt.tight_layout()

# Mark month‐boundaries on the x-axis
month_starts = [1,32,60,91,121,152,182,213,244,274,305,335]
month_names  = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.xticks(month_starts, month_names)
```

```python
import matplotlib.pyplot as plt
import calendar

# Convert to proper units
temp_C = df['Temperature'] / 10.0          # → °C

# Build a list of 12 series, one per month
monthly_temps = [temp_C[df.index.month == m] for m in range(1, 13)]

# Monthly Temperature
fig, ax = plt.subplots(figsize=(12, 6))
ax.boxplot(monthly_temps, tick_labels=calendar.month_abbr[1:], showfliers=False)
ax.set_title('Monthly Temperature Distribution (°C)')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (°C)')
ax.grid(True)
fig.tight_layout()
plt.show()
```

```python
# 1) Convert sunshine to hours: the raw unit is 0.1 h per hourly bin
sun_h = df['SunshineDuration'] / 10.0

# 2) Compute monthly totals
monthly = sun_h.resample('ME').sum()

# 3a) Bar chart for the climatology (all years)
fig, ax = plt.subplots(figsize=(10,5))
# extract month names
months = monthly.index.to_period('M').strftime('%b')
# group by month only
mon_clim = monthly.groupby(monthly.index.month).mean()
ax.bar(pd.Categorical(mon_clim.index.map(lambda m: pd.to_datetime(f'2000-{m:02d}-01').strftime('%b')),
                     categories=months.unique(), ordered=True),
       mon_clim.values)
ax.set_title('Climatological Monthly Sunshine Hours')
ax.set_xlabel('Month')
ax.set_ylabel('Average Total Sunshine (h)')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 3b) Grouped bars by year
# pivot to year×month
tab = monthly.to_frame('Sun_h').assign(
    Year=monthly.index.year,
    Month=monthly.index.month
).pivot(index='Month', columns='Year', values='Sun_h')

fig, ax = plt.subplots(figsize=(12,6))
tab.plot(kind='bar', ax=ax)
ax.set_xticklabels(pd.to_datetime(tab.index, format='%m').strftime('%b'), rotation=0)
ax.set_title('Monthly Sunshine Hours by Year')
ax.set_xlabel('Month')
ax.set_ylabel('Total Sunshine (h)')
ax.legend(title='Year', bbox_to_anchor=(1,1))
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 1. Compute daily total sunshine hours (convert 0.1 h units → hours)
daily_sun = (df['SunshineDuration'] / 10.0).resample('D').sum()

# 2. Build a helper DataFrame and extract month
daily_df = daily_sun.to_frame('DailySun_h')
daily_df['Month'] = daily_df.index.month

# 3. Gather per-month series for the boxplot
box_data = [daily_df[daily_df['Month'] == m]['DailySun_h'] for m in range(1, 13)]

# 4. Draw the boxplot
fig, ax = plt.subplots(figsize=(12, 6))
ax.boxplot(
    box_data,
    tick_labels=pd.to_datetime(range(1, 13), format='%m').strftime('%b')
)
ax.set_title('Monthly Distribution of Daily Sunshine Hours', fontsize=14, weight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Daily Sunshine Hours (h)', fontsize=12)
ax.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
